```python
# import the pandas module
# https://pandas.pydata.org/docs/user_guide/index.html

import pandas as pd

# numpy is a standard module for simple numerical calculations
import numpy as np

import scipy.stats
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data_frame = pd.read_csv("Census_T40.4.txt",delimiter="\t")
```

```python
data_frame=data_frame.replace("Unreliable",np.nan)
```

```python
data_frame[(data_frame["Census Region Code"] == "CENS-R1") & (data_frame["Gender"] == "Male") & 
           (data_frame["Notes"] != "Total")]
```

```python
CR1_M = data_frame[(data_frame["Census Region Code"] == "CENS-R1") & (data_frame["Gender"] == "Male") & 
           (data_frame["Notes"] != "Total")][["Year", "Crude Rate"]].astype(float)
CR1_F = data_frame[(data_frame["Census Region Code"] == "CENS-R1") & (data_frame["Gender"] == "Female") & 
           (data_frame["Notes"] != "Total")][["Year", "Crude Rate"]].astype(float)
CR2_M = data_frame[(data_frame["Census Region Code"] == "CENS-R2") & (data_frame["Gender"] == "Male") & 
           (data_frame["Notes"] != "Total")][["Year", "Crude Rate"]].astype(float)
CR2_F = data_frame[(data_frame["Census Region Code"] == "CENS-R2") & (data_frame["Gender"] == "Female") & 
           (data_frame["Notes"] != "Total")][["Year", "Crude Rate"]].astype(float)
CR3_M = data_frame[(data_frame["Census Region Code"] == "CENS-R3") & (data_frame["Gender"] == "Male") & 
           (data_frame["Notes"] != "Total")][["Year", "Crude Rate"]].astype(float)
CR3_F = data_frame[(data_frame["Census Region Code"] == "CENS-R3") & (data_frame["Gender"] == "Female") & 
           (data_frame["Notes"] != "Total")][["Year", "Crude Rate"]].astype(float)
CR4_M = data_frame[(data_frame["Census Region Code"] == "CENS-R4") & (data_frame["Gender"] == "Male") & 
           (data_frame["Notes"] != "Total")][["Year", "Crude Rate"]].astype(float)
CR4_F = data_frame[(data_frame["Census Region Code"] == "CENS-R4") & (data_frame["Gender"] == "Female") & 
           (data_frame["Notes"] != "Total")][["Year", "Crude Rate"]].astype(float)
```

```python
x1c=CR1_M["Year"]
#print(x1c)
```

```python
CR1_M_ante = CR1_M[CR1_M["Year"] != 2020]
CR1_M_covid = CR1_M[CR1_M["Year"] == 2020]
CR2_M_ante = CR2_M[CR2_M["Year"] != 2020]
CR2_M_covid = CR2_M[CR2_M["Year"] == 2020]
CR3_M_ante = CR3_M[CR3_M["Year"] != 2020]
CR3_M_covid = CR3_M[CR3_M["Year"] == 2020]
CR4_M_ante = CR4_M[CR4_M["Year"] != 2020]
CR4_M_covid = CR4_M[CR4_M["Year"] == 2020]
CR1_F_ante = CR1_F[CR1_F["Year"] != 2020]
CR1_F_covid = CR1_F[CR1_F["Year"] == 2020]
CR2_F_ante = CR2_F[CR2_F["Year"] != 2020]
CR2_F_covid = CR2_F[CR2_F["Year"] == 2020]
CR3_F_ante = CR3_F[CR3_F["Year"] != 2020]
CR3_F_covid = CR3_F[CR3_F["Year"] == 2020]
CR4_F_ante = CR4_F[CR4_F["Year"] != 2020]
CR4_F_covid = CR4_F[CR4_F["Year"] == 2020]

#print(CR4_F_covid)
```

```python
x1=CR1_M_ante["Year"]
y1=CR1_M_ante["Crude Rate"]
x2=CR1_F_ante["Year"]
y2=CR1_F_ante["Crude Rate"]

x3=CR2_M_ante["Year"]
y3=CR2_M_ante["Crude Rate"]
x4=CR2_F_ante["Year"]
y4=CR2_F_ante["Crude Rate"]

x5=CR3_M_ante["Year"]
y5=CR3_M_ante["Crude Rate"]
x6=CR3_F_ante["Year"]
y6=CR3_F_ante["Crude Rate"]

x7=CR4_M_ante["Year"]
y7=CR4_M_ante["Crude Rate"]
x8=CR4_F_ante["Year"]
y8=CR4_F_ante["Crude Rate"]
```

```python
CR1_M2=data_frame[(data_frame["Census Region Code"] == "CENS-R1") & (data_frame["Gender"] == "Male") & 
           (data_frame["Notes"] != "Total")]["Crude Rate"].astype(float)
CR1_F2=data_frame[(data_frame["Census Region Code"] == "CENS-R1") & (data_frame["Gender"] == "Female") & 
           (data_frame["Notes"] != "Total")]["Crude Rate"].astype(float)
CR2_M2=data_frame[(data_frame["Census Region Code"] == "CENS-R2") & (data_frame["Gender"] == "Male") & 
           (data_frame["Notes"] != "Total")]["Crude Rate"].astype(float)
CR2_F2=data_frame[(data_frame["Census Region Code"] == "CENS-R2") & (data_frame["Gender"] == "Female") & 
           (data_frame["Notes"] != "Total")]["Crude Rate"].astype(float)
CR3_M2=data_frame[(data_frame["Census Region Code"] == "CENS-R3") & (data_frame["Gender"] == "Male") & 
           (data_frame["Notes"] != "Total")]["Crude Rate"].astype(float)
CR3_F2=data_frame[(data_frame["Census Region Code"] == "CENS-R3") & (data_frame["Gender"] == "Female") & 
           (data_frame["Notes"] != "Total")]["Crude Rate"].astype(float)
CR4_M2=data_frame[(data_frame["Census Region Code"] == "CENS-R4") & (data_frame["Gender"] == "Male") & 
           (data_frame["Notes"] != "Total")]["Crude Rate"].astype(float)
CR4_F2=data_frame[(data_frame["Census Region Code"] == "CENS-R4") & (data_frame["Gender"] == "Female") & 
           (data_frame["Notes"] != "Total")]["Crude Rate"].astype(float)
Ratio_CR1 = CR1_M2.div(CR1_F2.values)
Ratio_CR2 = CR2_M2.div(CR2_F2.values)
Ratio_CR3 = CR3_M2.div(CR3_F2.values)
Ratio_CR4 = CR4_M2.div(CR4_F2.values)
```

```python
coeffM = np.polyfit(x1, y1, 2)
yfitM = np.polyval(coeffM,x1)
resM = np.sum((y1  - yfitM)**2)
coeffF = np.polyfit(x2, y2, 2)
yfitF = np.polyval(coeffF,x2)
resF = np.sum((y2  - yfitF)**2)

#stdM = np.sqrt(resM/5)
#stdF = np.sqrt(resF/5)

stdM = np.sqrt(resM/(len(y1)-3))
stdF = np.sqrt(resF/(len(y2)-3))


gdf = pd.DataFrame({"value": [2013, 2013.5, 2014, 2014.5, 2015, 2015.5, 2016, 2016.5, 2017, 
                              2017.5, 2018, 2018.5, 2019, 2019.5, 2020]})
```

```python
coeffM3 = np.polyfit(x3, y3, 2)
yfitM3 = np.polyval(coeffM3,x3)
resM3 = np.sum((y3  - yfitM3)**2)
coeffF4 = np.polyfit(x4, y4, 2)
yfitF4 = np.polyval(coeffF4,x4)
resF4 = np.sum((y4  - yfitF4)**2)

#stdM3 = np.sqrt(resM3/5)
#stdF4 = np.sqrt(resF4/5)

stdM3 = np.sqrt(resM3/(len(y3)-3))
stdF4 = np.sqrt(resF4/(len(y4)-3))
```

```python
coeffM5 = np.polyfit(x5, y5, 2)
yfitM5 = np.polyval(coeffM5,x5)
resM5 = np.sum((y5  - yfitM5)**2)
coeffF6 = np.polyfit(x6, y6, 2)
yfitF6 = np.polyval(coeffF6,x6)
resF6 = np.sum((y6  - yfitF6)**2)

#stdM5 = np.sqrt(resM5/5)
#stdF6 = np.sqrt(resF6/5)

stdM5 = np.sqrt(resM5/(len(y5)-3))
stdF6 = np.sqrt(resF6/(len(y6)-3))
```

```python
coeffM7 = np.polyfit(x7, y7, 2)
yfitM7 = np.polyval(coeffM7,x7)
resM7 = np.sum((y7  - yfitM7)**2)
coeffF8 = np.polyfit(x8, y8, 2)
yfitF8 = np.polyval(coeffF8,x8)
resF8 = np.sum((y8  - yfitF8)**2)

#stdM7 = np.sqrt(resM7/5)
#stdF8 = np.sqrt(resF8/5)

stdM7 = np.sqrt(resM7/(len(y7)-3))
stdF8 = np.sqrt(resF8/(len(y8)-3))
```
